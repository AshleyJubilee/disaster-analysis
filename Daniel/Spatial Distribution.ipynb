```python
df = pd.read_csv('disaster_data.csv', encoding='latin-1')

# Specify the columns you want to keep
selected_columns = [
    'Disaster Group', 'Disaster Subgroup', 'Disaster Type', 'Disaster Subtype',
    'ISO', 'Country', 'Subregion', 'Region', 'Location',
    'Magnitude', 'Magnitude Scale', 'Latitude', 'Longitude',
    'Start Year', 'Start Month',
    'End Year',
    'Total Deaths',
    'Total Damage, Adjusted (\'000 US$)'
]

# Select the relevant columns
selected_df = df[selected_columns]

# Group by 'Disaster Type' and count occurrences
disaster_type_counts = selected_df.groupby('Disaster Type').size().reset_index(name='Frequency')

# Sort by frequency in descending order
disaster_type_counts = disaster_type_counts.sort_values(by='Frequency', ascending=False)

# Display the most common disaster types
print(disaster_type_counts)
```

```python
selected_df = df[selected_columns]

# Group by 'Disaster Type' and count occurrences
disaster_type_counts = selected_df['Disaster Type'].value_counts().reset_index(name='Frequency')

# Sort by frequency in descending order
disaster_type_counts = disaster_type_counts.sort_values(by='Frequency', ascending=False)

# Take the top 10 most common disaster types
top_10_disasters = disaster_type_counts.head(10)

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_10_disasters['Disaster Type'], top_10_disasters['Frequency'], color='skyblue')
plt.xlabel('Disaster Type')
plt.ylabel('Frequency')
plt.title('Top 10 Most Common Disaster Types Globaly')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# Display unique regions in the entire DataFrame
unique_regions = df['Region'].unique()
print("Unique Regions:", unique_regions)

# Filter the data for Africa region
selected_df_africa = df[df['Region'] == 'Africa']

# Filter the data for Natural disasters
selected_df_natural_africa = selected_df_africa[selected_df_africa['Disaster Group'] == 'Natural']

# Create a bar plot for the top 10 natural disaster subtypes in Africa
plt.figure(figsize=(10, 6))
bar_plot = sns.countplot(x='Disaster Subtype', data=selected_df_natural_africa, palette='viridis',
                         order=selected_df_natural_africa['Disaster Subtype'].value_counts().index[:10])

# Customize the appearance of the bars
bar_plot.set_xticklabels(bar_plot.get_xticklabels(), rotation=45, ha="right")
bar_plot.set_title('Top 10 Natural Disaster Subtypes in Africa')
bar_plot.set_xlabel('Disaster Subtype')
bar_plot.set_ylabel('Count')

plt.tight_layout()
plt.show()
```
